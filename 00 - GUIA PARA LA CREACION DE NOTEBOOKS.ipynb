# Guía general para los notebooks del curso

Se sugiere que los notebooks del repositorio tengan sigan la siguiente estructura

- Los notebooks se numeran `00 -`, `01 -`, etc. como caracteres iniciales de su propio nombre, o con algún esquema numérico de modo que queden ordenados al listarlos.
